```python
# Summary of missing values
df.isnull().sum()
```

```python
# Check unique values in important columns
print("Mechanics:", df['mechanic'].unique())
print("Category:", df['category'].unique())
print("Force:", df['force'].unique())
print("Equipment:", df['equipment'].unique())
print("Level:", df['level'].unique())
```

```python
# Check if stringified lists need parsing
print(df['instructions'].iloc[0])
print(df['primaryMuscles'].iloc[0])
```

```python
# Clean missing values
df['force'] = df['force'].fillna('unspecified')
df['mechanic'] = df['mechanic'].fillna('unspecified')
df['equipment'] = df['equipment'].fillna('unspecified')

# Standardise casing
for col in ['force', 'mechanic', 'equipment', 'category', 'level']:
    df[col] = df[col].str.lower()

# Convert stringified lists to actual Python lists
import ast
list_cols = ['primaryMuscles', 'secondaryMuscles', 'instructions', 'images']
for col in list_cols:
    df[col] = df[col].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else [])
```

```python
print("Summary of missing values after cleaning")
df.isnull().sum()
```

```python
# save clean dataset 
df.to_csv('clean_exercises.csv', index=False)
```
